***Label Encoding***<br>
It is a popular encoding technique for handling categorical variables. In this technique, each label is assigned a unique integer based on alphabetical ordering. Let's see how to implement label encoding in Python using the scikit-learn library and also understand the challenges with label encoding.

***its apply on the States feature as they are categorical values.***

```python
from sklearn.preprocessing import LabelEncoder
lab_encoding = LabelEncoder()
df.State = lab_encoding.fit_transform(df.State)
df.State.unique()
```

***from X rows its remove the Profit value and on y in one Profit single feature present as(dependent and independent features).***

```python
X = df.drop(['Profit'], axis=1)
y = df['Profit']
```

***Splitting the dataset into Training and Testing set***

```python
#Splitting the dataset into Training-set and Testing-set
import sklearn
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
print("Shape of X_train",X_train.shape)
print("Shape of X_test",X_test.shape)
print("Shape of y_train",y_train.shape)
print("Shape of y_test",y_test.shape)
```

***Fitting Simple Linear Regression to the Training Dataset***

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

***Evaluation of model / Prediction of test Dataset***

```python
# Predicting the values (Scores)
y_pred=model.predict(X_test)
y_pred
```

***finding the score from the given dataset***

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```

***finding the mean squared error***

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
mse
```
